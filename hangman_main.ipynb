```python
import string
import random
from hangman_module import words
from hangman_visual import lives_visual_dict

#mengambil kata dari library
def get_valid_word(words):
    word = random.choice(words)
    while '-' in word or ' ' in word:
        word = random.choice(words)
    
    return word.upper()

#fungsi hangman
def hangman():
    word = get_valid_word(words)
    word_letters = set(word)
    alphabet = set(string.ascii_uppercase)
    used_letters = set()
    
    lives = 7
    
#tebak kata, print gambar jika huruf tidak termasuk dalam kata yang harus ditebak
    while len(word_letters) > 0 and lives > 0:
        print('Sisa nyawa: ', lives, 'Huruf yang salah: ', ''.join(used_letters))
        
        word_list = [letter if letter in used_letters else '-' for letter in word]
        print(lives_visual_dict[lives])
        print('Huruf yang ditebak: ', ''.join(word_list))
        
        user_letter = input('Tebak kata: ').upper()
        if user_letter in alphabet - used_letters:
            used_letters.add(user_letter)
            if user_letter in word_letters:
                word_letters.remove(user_letter)
                print('')
#nyawa akan berkurang jika salah menebak huruf           
            else:
                lives = lives - 1
                print('\nHuruf', user_letter, ' tidak ada di kata.')

        elif user_letter in used_letters:
            print('/nHuruf sudah dipakai. Coba Lagi.')

        else:
            print('/nHuruf tidak valid. Coba Lagi.')
            
    if lives == 0:
        print(lives_visual_dict[lives])
        print('Oops You Lose. Kata yang ditebak adalah ', word)
    else:
        print('Congrats, kamu berhasil menebak kata ', word, ' !!!')
        
if __name__ == '__main__':
    hangman()
```
